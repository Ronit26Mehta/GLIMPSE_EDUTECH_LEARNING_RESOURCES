 The number of solar heating systems available to the public is quite large, and their heat-storage capacities are quite varied. Here is a distribution of heat-storage capacity (in days) of 28 systems that were tested recently by University Laboratories, Inc.: 
 
University Laboratories, Inc., knows that its report on the tests will be widely circulated and used as the basis for tax legislation on solar-heat allowances. It therefore wants the measures it uses to be as reflective of the data as possible. 
(a) Compute the mean for these data. 
(b) Compute the mode for these data. 
(c) Compute the median for these data. 
(d) Select the answer among parts (a), (b), and (c) that best reflects the central tendency of the test data and justify your choice.


```python
days_midpoint = np.array([0.5,1.5,2.5,3.5,4.5,5.5,6.5])
frequency = np.array([2,4,6,7,5,3,1])
mean_value = np.average(days_midpoint,weights=frequency)
mode_value_1 = np.repeat(days_midpoint,frequency)
mode_value = mode(mode_value_1)
mode_value = mode_value[0]
cumulative_freq = np.cumsum(frequency)
median_index = np.argmax(cumulative_freq >= (np.sum(frequency)/2))
median_value = np.median(days_midpoint[median_index:median_index+2])

print("the result is mean:{},median:{},mode:{}".format(mean_value,median_value,mode_value))
```

15. A company is considering employing one of two training programs. Two groups were trained for the same task. Group 1 was trained by program A; group 2, by program B. For the first group, the times required to train the employees had an average of 32.11 hours and a variance of 68.09. In the second group, the average was 19.75 hours and the variance was 71.14. Which training program has less relative variability in its performance?

```python
mean_1 = 32.11
variance_1 = 68.09
mean_2 = 19.75
variance_2=71.14
std_1 = np.sqrt(variance_1)
std_2 = np.sqrt(variance_2)
cov_1 = (std_1/mean_1) * 100
cov_2 = (std_2/mean_2) * 100
if cov_1 > cov_2 :
    print("2 has less reliabilty")
elif cov_1 < cov_2:
    print("1 has less reliabilty")
else :
    print("not reliable")
print(cov_1)
print(cov_2)
```

16. Here is a frequency distribution of the weight of 150 people who used a ski lift a certain day. Construct a histogram for these data.
 
(a) What can you see from the histogram about the data that was not immediately apparent from the frequency distribution? 
(b) If each ski lift chair holds two people but is limited in total safe weight capacity to 400 pounds, what can the operator do to maximize the people capacity of the ski lift without exceeding the safe weight capacity of a chair? Do the data support your proposal?


```python
import matplotlib.pyplot as plt

class_intervals = ['75-89', '90-104', '105-119', '120-134', '135-149', '150-164', '165-179', '180-194', '195-209', '210-224']
frequencies = [10, 11, 23, 36, 31, 23, 9, 9, 6, 2]

plt.figure(figsize=(10, 6))
plt.bar(class_intervals, frequencies, color='skyblue', edgecolor='black', linewidth=1)
plt.title('Histogram of Weight Distribution')
plt.xlabel('Weight Class Intervals')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

```

```python
max_people_capacity = 0
total_weight_capacity = 400
for interval, freq in zip(class_intervals, frequencies):
    weight_range = [int(x) for x in interval.split('-')]
    mid_weight = (weight_range[0] + weight_range[1]) / 2
    people_capacity = min(freq, total_weight_capacity // (mid_weight * 2))
    max_people_capacity += people_capacity
```

```python
print(max_people_capacity)
```

17. Test scores for a college statistics class held during the day are:
99 56 78 55.5 32 90 80 81 56 59 45 77 84.5 84 70 72 68 32 79 90

98 78 68 83 81 89 88 76 65 45 98 90 80 84.5 85 79 78 98 90 79 81 25.5

a.	Find the smallest and largest values, the median, and the first and third quartile for the day class.

c.	For each data set, what percentage of the data is between the smallest value and the first quartile? the first quartile and the median? the median and the third quartile? the third quartile and the largest value? What percentage of the data is between the first quartile and the largest value?
d.	Create a box plot for each set of data. Use one number line for both box plots.
e.	Which box plot has the widest spread for the middle 50% of the data (the data between the first and third quartiles)? What does this mean for that set of data in comparison to the other set of data.


```python
day_stats = np.array([99, 56, 78, 55.5, 32, 90, 80, 81, 56, 59, 45, 77, 84.5, 84, 70, 72, 68, 32, 79, 90])
night_stats = np.array([98, 78, 68, 83, 81, 89, 88, 76, 65, 45, 98, 90, 80, 84.5, 85, 79, 78, 98, 90, 79, 81, 25.5])
print(np.max(day_stats),np.min(day_stats),np.quantile(day_stats,q=0.25),np.quantile(day_stats,q=0.75))
print(np.max(night_stats),np.min(night_stats),np.quantile(night_stats,q=0.25),np.quantile(night_stats,q=0.75))
```

```python
import numpy as np

day_class_scores = np.array([99, 56, 78, 55.5, 32, 90, 80, 81, 56, 59, 45, 77, 84.5, 84, 70, 72, 68, 32, 79, 90])
night_class_scores = np.array([98, 78, 68, 83, 81, 89, 88, 76, 65, 45, 98, 90, 80, 84.5, 85, 79, 78, 98, 90, 79, 81, 25.5])

def percentage_between(arr, start, end):
    count = np.sum((arr >= start) & (arr <= end))
    total = len(arr)
    percentage = (count / total) * 100
    return percentage

percentage_day_q1_median = percentage_between(day_class_scores, np.min(day_class_scores), np.percentile(day_class_scores, 25))
percentage_day_median_q3 = percentage_between(day_class_scores, np.percentile(day_class_scores, 25), np.percentile(day_class_scores, 50))
percentage_day_q3_max = percentage_between(day_class_scores, np.percentile(day_class_scores, 50), np.max(day_class_scores))
percentage_day_q1_max = percentage_between(day_class_scores, np.percentile(day_class_scores, 25), np.max(day_class_scores))
percentage_day_q1_max_total = percentage_between(day_class_scores, np.percentile(day_class_scores, 25), np.max(day_class_scores))

percentage_night_q1_median = percentage_between(night_class_scores, np.min(night_class_scores), np.percentile(night_class_scores, 25))
percentage_night_median_q3 = percentage_between(night_class_scores, np.percentile(night_class_scores, 25), np.percentile(night_class_scores, 50))
percentage_night_q3_max = percentage_between(night_class_scores, np.percentile(night_class_scores, 50), np.max(night_class_scores))
percentage_night_q1_max = percentage_between(night_class_scores, np.percentile(night_class_scores, 25), np.max(night_class_scores))
percentage_night_q1_max_total = percentage_between(night_class_scores, np.percentile(night_class_scores, 25), np.max(night_class_scores))

print(percentage_day_q1_median)
print(percentage_day_median_q3)
print(percentage_day_q3_max)
print(percentage_day_q1_max)
print(percentage_day_q1_max_total)

print(percentage_night_q1_median)
print(percentage_night_median_q3)
print(percentage_night_q3_max)
print(percentage_night_q1_max)
print(percentage_night_q1_max_total)
```
